```python
# importing required modules
from PyPDF2 import PdfMerger,PdfWriter, PdfReader
import pyttsx3 
import fitz  # PyMuPDF

def pdf_merger():
    
    pdfs = []
    n = int(input("Enter the number of pdfs to be merged: "))
    for i in range(1,n+1):
        pdf_name = input(f"Enter the name of the pdf{i}: ")
        pdfs.append(pdf_name)

    merger = PdfMerger()

    for pdf in pdfs:
            merger.append(pdf)

    merger.write(input("Enter the name of output PDF: "))
    merger.close()

def pdf_rotation():
    
    pdf_file =input("Enter the input file: ")
    output_file = input("Enter the name of output file: ")
    reader = PdfReader(pdf_file)
    writer = PdfWriter()

    print(len(reader.pages))
    rotation_angle = (int(input("Enter the angle of rotation (multiple of 90): ")))

    option=int(input("Enter the choice\n1.Rotate complete pdf\n2.Rotate particular page\n3.Rotate a page and retain the remaining pages\n4.Rotate a particular range of pages\n  "))

    if(option==1):       #complete rotation
        for i in range(0,len(reader.pages)):
            writer.add_page(reader.pages[i])
            writer.pages[i].rotate(rotation_angle)

    elif(option==2):      #particular pages
        page_number=int(input("Enter the number of the page to rotated: "))
        writer.add_page(reader.pages[page_number-1])
        writer.pages[0].rotate(rotation_angle)

    elif(option==3):        #only a particular page
        page_number=int(input("Enter the page number: "))
        if (page_number>0 and page_number<=len(reader.pages)):
            for i in range(len(reader.pages)):
                page = reader.pages[i]
                if i == page_number-1:
                    page.rotate(rotation_angle)
                writer.add_page(page)

    elif(option==4):      #particular range
        start_page=int(input("Enter the start page:"))
        end_page=int(input("Enter the end page: "))
        for i in range(len(reader.pages)):
            page = reader.pages[i]
            if start_page <= i + 1 <= end_page:
                page.rotate(rotation_angle)
            writer.add_page(page)
    

    with open(output_file, "wb") as out:
        writer.write(out)    

def pdf_splitter():

    pdf_document= input("Enter the PDF name: ")

    pdf_reader = PdfReader(pdf_document)

    pdf_writer = PdfWriter()

    length = len(pdf_reader.pages)


    print("Enter the split range of pdf from 1 to",length)

    #split range
    pdf_start=int(input("Enter start page: "))  #start
    pdf_end=int(input("Enter end page: "))          #end

    new_pdf=input("Enter the name of the new PDF:  ")

    

    for page in range(pdf_start-1,pdf_end):
        current_page = pdf_reader.pages[page]
        pdf_writer.add_page(current_page)


    with open(new_pdf, "wb") as out:
        pdf_writer.write(out)

def pdf_watermark():
    
    input_pdf=input("Enter the input file: ")
    output=input("Enter the name of the output file: ")
    watermark=input("Enter the name of the watermarked file: ")
    watermark_file = PdfReader(watermark)
    watermark_page = watermark_file.pages[0]

    pdf_reader = PdfReader(input_pdf)
    pdf_writer = PdfWriter()

        # Watermark all the pages
    for page in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page]
        page.merge_page(watermark_page)
        pdf_writer.add_page(page)

    with open(output, 'wb') as out:
        pdf_writer.write(out)    

def pdf_to_audio():
     
    # path of the PDF file 
    pdf_name = input("Enter the file name: ")
    path = open(pdf_name, 'rb') 

    # creating a PdfFileReader object 
    pdf_Reader =PdfReader(path) 

    print("Enter the page numbers between 1 to",len(pdf_Reader.pages))

#     the page with which you want to start and end
    page_number = int(input('Enter the starting page number: '))
    end_page = int(input("Enter the end page number:"))


    for i in range(page_number-1,end_page):
        print(i)
        from_page = pdf_Reader.pages[i]

        # extracting the text from the PDF 
        text = from_page.extract_text() 

        # reading the text 
        speak = pyttsx3.init() 
        speak.say(text) 
        speak.runAndWait()    

def pdf_text():
    print("1.print text of a page\n2.print text of a paragraph\n")
    choice=int(input("enter your choice: "))

    if(choice==1):
             
        pdf_file =input("\nEnter the pdf file name : ")
        # creating a pdf reader object
        reader = PdfReader(pdf_file)
    
        # printing number of pages in pdf file
        print("Total number of pages: ",len(reader.pages))
        print("Enter the number of the page to be printed 1 to ",len(reader.pages))
        n=int(input())
        if n>0 and n<=len(reader.pages):
            # getting a specific page from the pdf file
            page = reader.pages[n-1]

            # extracting text from page
            text = page.extract_text()
            print(text)
        else:
            print("Invalid page number") 
    elif(choice==2):
        

        pdf_path = input("Enter the name of the pdf: ")
        pdf_document = fitz.open(pdf_path)
        print("Total number of pages in PDF: ",pdf_document.page_count)

        page_number = int(input("Enter the page number: "))-1 
        if(page_number>0 and page_number<pdf_document.page_count):

            paragraph_number = int(input("Enter the paragraph number: "))-1
       
            page = pdf_document[page_number]

            blocks = page.get_text("blocks")
            non_empty_blocks = [block for block in blocks if block[4].strip()]
            selected_block = non_empty_blocks[paragraph_number] if paragraph_number < len(non_empty_blocks) else None

            if selected_block:
                text = selected_block[4]  # 5th element contains the text of the block
            else:
                text = "Paragraph not found on the specified page."
           
            print(text)
        else:
            print("Invalid page number")
        pdf_document.close() 
        
while(1):
    print("Select an operation to be performed on PDF:")
    print("\t1:- Extract text")
    print("\t2:- Merge PDFs")
    print("\t3:- Apply watermark")
    print("\t4:- split pdf ")    
    print("\t5:- Rotating a PDF ")
    print("\t6:- text to audio")
    print("\t7:- exit")
    option = int(input("enter the option: "))
    if(option==1):
        pdf_text()

    elif(option==2):
        pdf_merger()

    elif(option==3):
        pdf_watermark()

    elif(option==4):
        pdf_splitter()

    elif(option==5):
        pdf_rotation()

    elif(option==6):
        pdf_to_audio()

    else:
        break
```
